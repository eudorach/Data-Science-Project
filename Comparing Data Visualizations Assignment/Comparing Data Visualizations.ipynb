```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('countries.csv')
```

```python
df.head()
```

```python
df_1952 = df[df.year==1952]
```

```python
df_1952
```

```python
df_1952_africa = df_1952[df_1952.continent=='Africa']
```

```python
df_1952_africa
```

```python
df_1952_asia = df_1952[df_1952.continent == 'Asia']
```

```python
df_1952_asia.head()
```

```python
plt.hist(df_1952_africa.lifeExpectancy, 20)
```

```python
plt.hist(df_1952_asia.lifeExpectancy, 20)
```

```python
plt.subplot(2,1,1)
plt.title('Distribution of Life Expectancy in 1952')
plt.hist(df_1952_africa.lifeExpectancy, 20, range = (25,))
plt.ylabel('Africa')
plt.subplot(2,1,2)
plt.hist(df_1952_asia.lifeExpectancy, 20)
plt.ylabel('Asia')
```
