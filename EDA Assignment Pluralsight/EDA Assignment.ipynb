### Univariate Analysis

```python
diamonds_color = diamonds.groupby('color')
```

```python
diamonds_e = diamonds[diamonds['color'] == 'E']

diamonds_e.reset_index(inplace=True)

diamonds_e.head()

#making a DF that only contains the color E 
```

```python
diamonds_f = diamonds[diamonds['color'] == 'F']

diamonds_f.reset_index(inplace=True)

diamonds_f.head()

#making a DF that only contains the color F
```

```python
stats.ttest_ind(diamonds_e['carat'], diamonds_f['carat']) 

#null hypothesis: the mean carats of diamonds with color E is the same as the mean carat of diamonds with color F

#the result: reject null hypothesis; the mean carats of color E and F are not the same
```

```python
fig, ax = plt.subplots(figsize = (15,8))

sns.scatterplot(x='carat', y='price', data = diamonds, color='#b3b3ff', ax=ax)

#scatter plot to show the relationship between carat value and price of the diamonds
```

```python
x = diamonds['carat']

y = diamonds['price']

a, b = np.polyfit(x,y,1)
```

```python
plt.scatter(x,y,color='#b3b3ff')
plt.title("The relationship between carat and price of the diamonds")
plt.xlabel("Carats")
plt.ylabel("Price in USD")
plt.show()
plt.plot(x, a*x+b)
```
